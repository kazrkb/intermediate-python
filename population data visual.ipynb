```python
# Country data with GDP, Life Expectancy, Population, and Continent
countries = ['China', 'India', 'United States', 'Indonesia', 'Pakistan', 
             'Brazil', 'Nigeria', 'Bangladesh', 'Russia', 'Mexico',
             'Japan', 'Philippines', 'Ethiopia', 'Vietnam', 'Egypt',
             'Germany', 'Turkey', 'Iran', 'Thailand', 'United Kingdom',
             'France', 'Italy', 'South Africa', 'Myanmar', 'South Korea',
             'Colombia', 'Spain', 'Ukraine', 'Argentina', 'Algeria',
             'Sudan', 'Uganda', 'Iraq', 'Afghanistan', 'Poland',
             'Canada', 'Morocco', 'Saudi Arabia', 'Uzbekistan', 'Peru']

# GDP per capita (in thousands USD)
gdp_per_capita = [10.5, 2.3, 63.5, 4.1, 1.5, 
                  8.9, 2.1, 2.1, 11.6, 9.9,
                  40.1, 3.5, 0.9, 3.8, 3.0,
                  46.3, 9.1, 2.4, 7.8, 42.3,
                  39.0, 31.3, 6.4, 1.4, 31.8,
                  6.1, 27.0, 3.7, 10.0, 4.0,
                  0.8, 0.8, 4.2, 0.5, 15.7,
                  46.3, 3.2, 23.1, 1.7, 6.7]

# Convert GDP to log scale for better visualization
import numpy as np
gdp_log = [np.log10(gdp) for gdp in gdp_per_capita]

# Life expectancy (in years)
life_expectancy = [76.9, 69.7, 78.9, 71.7, 67.3,
                   75.9, 54.7, 72.6, 72.6, 75.1,
                   84.6, 71.2, 66.2, 75.4, 71.9,
                   81.3, 77.7, 76.7, 77.2, 81.3,
                   82.7, 83.6, 64.1, 67.1, 83.0,
                   77.3, 83.6, 72.1, 76.7, 76.9,
                   65.3, 63.0, 70.6, 64.8, 78.7,
                   82.4, 76.8, 75.1, 71.2, 76.7]

# Population (in billions)
population = [1.412, 1.408, 0.331, 0.273, 0.231, 
              0.215, 0.218, 0.166, 0.146, 0.128,
              0.125, 0.110, 0.120, 0.098, 0.104,
              0.083, 0.084, 0.084, 0.070, 0.067,
              0.068, 0.059, 0.059, 0.054, 0.052,
              0.051, 0.047, 0.044, 0.045, 0.044,
              0.046, 0.047, 0.041, 0.039, 0.038,
              0.038, 0.037, 0.035, 0.034, 0.033]

# Continents
continents = ['Asia', 'Asia', 'North America', 'Asia', 'Asia',
              'South America', 'Africa', 'Asia', 'Europe', 'North America',
              'Asia', 'Asia', 'Africa', 'Asia', 'Africa',
              'Europe', 'Asia', 'Asia', 'Asia', 'Europe',
              'Europe', 'Europe', 'Africa', 'Asia', 'Asia',
              'South America', 'Europe', 'Europe', 'South America', 'Africa',
              'Africa', 'Africa', 'Asia', 'Asia', 'Europe',
              'North America', 'Africa', 'Asia', 'Asia', 'South America']

# Create color map for continents
continent_colors = {'Asia': 'red', 'North America': 'blue', 'South America': 'green', 
                    'Europe': 'orange', 'Africa': 'purple'}

# Create bubble chart
plt.figure(figsize=(15, 10))

# Plot each continent separately for legend
for continent in continent_colors.keys():
    mask = [c == continent for c in continents]
    x_cont = [gdp_log[i] for i, m in enumerate(mask) if m]  # Use log GDP values
    y_cont = [life_expectancy[i] for i, m in enumerate(mask) if m]
    s_cont = [population[i]*16000 for i, m in enumerate(mask) if m]  # Scale bubble size (reduced by 5x)
    
    # Plot without label first for the actual data
    plt.scatter(x_cont, y_cont, s=s_cont, alpha=0.7, 
               c=continent_colors[continent])

# Create separate legend entries with uniform size
legend_elements = []
for continent, color in continent_colors.items():
    legend_elements.append(plt.scatter([], [], s=50, c=color, alpha=0.7, label=continent))

# Add country labels
for i, country in enumerate(countries):
    plt.annotate(country, (gdp_log[i], life_expectancy[i]), 
                xytext=(5, 5), textcoords='offset points', fontsize=8)

# Customize the plot
plt.xlabel("GDP per Capita (Log10 scale)")
plt.ylabel("Life Expectancy (years)")
plt.title("GDP vs Life Expectancy by Country\n(Bubble Size = Population, GDP on Log Scale)")
plt.legend(title="Continent")  # Now uses uniform legend bubbles
plt.grid(True, alpha=0.3)

# Adjust layout
plt.tight_layout()
plt.show()
```

# Analysis: GDP vs Life Expectancy Bubble Chart

## Key Findings from the Data Visualization

### 1. **Strong Positive Correlation Between GDP and Life Expectancy**
- Countries with higher GDP per capita consistently show higher life expectancy
- This relationship follows a **logarithmic pattern** - initial increases in GDP have larger impacts on life expectancy than later increases
- The relationship demonstrates **diminishing returns**: going from $1k to $10k GDP has a bigger health impact than going from $40k to $50k

### 2. **Continental Development Patterns**

#### 🟠 **Europe (Orange)**
- **Highest cluster**: Most European countries are in the upper-right quadrant
- High GDP ($15k-$60k) and high life expectancy (78-84 years)
- Examples: Germany, France, Italy, Spain, UK

#### 🔴 **Asia (Red)** 
- **Widest variation**: Shows the full spectrum of development
- Ranges from very poor (Afghanistan: $0.5k GDP, 65 years) to very rich (Japan: $40k GDP, 85 years)
- Large population countries (China, India) have moderate GDP but huge bubbles

#### 🟣 **Africa (Purple)**
- **Development challenges**: Concentrated in lower GDP and life expectancy regions
- Most countries below $10k GDP and under 75 years life expectancy
- Shows the need for continued development focus

#### 🔵 **North America (Blue)**
- **High development**: US and Canada both show high GDP and life expectancy
- Mexico shows middle-income characteristics

#### 🟢 **South America (Green)**
- **Middle-income range**: Generally in the middle for both metrics
- Countries like Brazil, Argentina, Colombia cluster together

### 3. **Population Size Insights**
- **Large populations ≠ High individual wealth**: China and India have the biggest bubbles but moderate GDP per capita
- **Small wealthy nations**: Countries like Germany, Japan have smaller populations but high individual prosperity
- **Population size varies independently** of economic development level

### 4. **Development Economics Principles Demonstrated**
- **Preston Curve**: The classic relationship between national income and life expectancy is clearly visible
- **Health-Wealth Nexus**: Economic prosperity strongly correlates with health outcomes
- **Continental Development Gaps**: Clear differences in development levels between continents

### 5. **Policy Implications**
- Economic growth policies can have significant health benefits, especially for lower-income countries
- Continental cooperation and development aid could help close health and prosperity gaps
- Large population countries face unique challenges in raising per-capita prosperity
