```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
year = [1950, 1960, 1970, 1980, 1990, 2000, 2010, 2020, 2030]
pop = [2.532, 2.934, 3.239, 4.856, 5.786, 6.954, 8.674, 10.673, 13.113]

plt.plot(year, pop)
plt.xlabel("Year")
plt.ylabel("Population")
plt.title("World Population Projections")
plt.show()
```

```python
year = [1950, 1960, 1970, 1980, 1990, 2000, 2010, 2020, 2030]
pop = [2.532, 2.934, 3.239, 4.856, 5.786, 6.954, 8.674, 10.673, 13.113]
# Add more data
year = [1920, 1930, 1940] + year
pop = [1.523, 1.845, 2.223] + pop


plt.plot(year, pop)
plt.xlabel("Year")
plt.ylabel("Population")
plt.title("World Population Projections")
plt.yticks([0, 2, 4, 6, 8, 10, 12, 14],
           ['0', '2B', '4B', '6B','8B','10B','12B','14B'])
plt.grid()
plt.show()
```
